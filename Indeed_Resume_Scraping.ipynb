```python
import numpy as np
from bs4 import BeautifulSoup
import urllib
import re
import requests
from urllib.request import urlopen
import pandas as pd

#Scrape first 50 resumes data from specified URL
import requests
s = requests.Session()
s.headers.update({'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'})
#k = s.get("https://resumes.indeed.com/search?q=java&l=&searchFields=jt")
#https://resumes.indeed.com/search?q=data+analyst&l=&searchFields=jt
r = s.get('https://resumes.indeed.com/rpc/search?l=&q=data+analyst')
result_json = r.json()

#Searching through first 50 resumes for work experience information
all_resumes_information_list = []
for resume_data in result_json['results']:
    r_url= "https://resumes.indeed.com/resume/"+resume_data['accountKey']
    m = requests.Session()
    m.headers.update({'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'})
    n = m.get(r_url)
    r_page= n.text
    
    person_data = []
    for i in range(0,5):
        
        titles_list = r_page.split("\\x22title\\x22:\\x22")
        try:
            thisTitle = titles_list[i+1].split("\\")[0]
            person_data.append(thisTitle)
            #gotdata = thisTitle[i]
        except (IndexError, ValueError):
            thisTitle = 'null'
        
        company_list = r_page.split("\\x22company\\x22:\\x22")
        try:
            thisCompany = company_list[i+1].split("\\")[0]
            person_data.append(thisCompany)
        except (IndexError, ValueError):
            thisCompany = 'null'
        
        date_list = r_page.split("\\x22dateRange\\x22:\\x22")
        try:
            thisDate = date_list[i+1].split("\\")[0]
            person_data.append(thisDate)
        except (IndexError, ValueError):
            thisDate = 'null'
            
        
    all_resumes_information_list.append(person_data)

#Scaping second page of 50 Resumes to attach to first set of 50
s2 = requests.Session()
s2.headers.update({'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'})
#k2 = s2.get("https://resumes.indeed.com/search?l=&q=Java&searchFields=jt&start=50")
#search?l=&q=data%20analyst&searchFields=jt&start=50
r2 = s2.get('https://resumes.indeed.com/rpc/search?l=&q=data+analyst&start=50')
result_json2 = r2.json()

#Search through second set of resumes for experience information
all_resumes_information_list2 = []
for resume_data2 in result_json2['results']:
    r_url2= "https://resumes.indeed.com/resume/"+resume_data2['accountKey']
    m2 = requests.Session()
    m2.headers.update({'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'})
    n2 = m2.get(r_url2)
    r_page2= n2.text
    
    person_data2 = []
    for j in range(0,5):
        
        titles_list = r_page2.split("\\x22title\\x22:\\x22")
        try:
            thisTitle = titles_list[j+1].split("\\")[0]
            person_data2.append(thisTitle)
            #gotdata = thisTitle[i]
        except (IndexError, ValueError):
            thisTitle = 'null'
        
            
        company_list = r_page2.split("\\x22company\\x22:\\x22")
        try:
            thisCompany = company_list[j+1].split("\\")[0]
            person_data2.append(thisCompany)
        except (IndexError, ValueError):
            thisCompany = 'null'
        
        date_list = r_page2.split("\\x22dateRange\\x22:\\x22")
        try:
            thisDate = date_list[j+1].split("\\")[0]
            person_data2.append(thisDate)
        except (IndexError, ValueError):
            thisDate = 'null'
               
    all_resumes_information_list2.append(person_data2)

#Combine resumes lists 1 and 2
experience_100 = all_resumes_information_list + all_resumes_information_list2
#print(experience_100[1])

#Continue searching through the resumes to adding education location and degree earned
education_list = []
for resume_data in result_json['results']:
    r_url= "https://resumes.indeed.com/resume/"+resume_data['accountKey']
    m = requests.Session()
    m.headers.update({'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'})
    n = m.get(r_url)
    r_page= n.text
    
    person_edu = []
    for i in range(0,2):

        deg_list = r_page.split("\\x22degree\\x22:\\x22")
        try:
            thisDeg = deg_list[i+1].split("\\")[0]
            person_edu.append(thisDeg)
        except (IndexError, ValueError):
            thisDeg = 'null'

        edu_list = r_page.split("\\x22university\\x22:\\x22")
        try:
            thisEdu = edu_list[i+1].split("\\")[0]
            person_edu.append(thisEdu)
        except (IndexError, ValueError):
            thisEdu = 'null'
    
    education_list.append(person_edu)

#Second set of 50 resumes for education location and degree
education_list2 = []
for resume_data in result_json2['results']:
    r_url= "https://resumes.indeed.com/resume/"+resume_data['accountKey']
    m = requests.Session()
    m.headers.update({'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'})
    n = m.get(r_url2)
    r_page= n.text
    
    person_edu2 = []
    for i in range(0,2):

        deg_list = r_page.split("\\x22degree\\x22:\\x22")
        try:
            thisDeg = deg_list[i+1].split("\\")[0]
            person_edu2.append(thisDeg)
        except (IndexError, ValueError):
            thisDeg = 'null'

        edu_list = r_page.split("\\x22university\\x22:\\x22")
        try:
            thisEdu = edu_list[i+1].split("\\")[0]
            person_edu2.append(thisEdu)
        except (IndexError, ValueError):
            thisEdu = 'null'
    
    education_list2.append(person_edu2)

#Combine education lists 1 and 2
edu_100 = education_list + education_list2
#print(edu_100[1])

#Loading experience attributes into dataframe
experience = pd.DataFrame(experience_100)
experience.rename(columns={0: 'work_title1', 1: 'work_company1', 2: 'work_dates1',
                          3: 'work_title2', 4: 'work_company2', 5: 'work_dates2',
                          6: 'work_title3', 7: 'work_company3', 8: 'work_dates3',
                          9: 'work_title4', 10: 'work_company4', 11: 'work_dates4',
                          12: 'work_title5', 13: 'work_company5', 14: 'work_dates5'}, inplace = True)

#Loading education attributes into dataframe
education = pd.DataFrame(edu_100)
education.rename(columns={0: 'degree1', 1: 'univeristy1',
                          2: 'degree2', 3: 'univeristy2'}, inplace = True)

#Combine experience dataframe and education dataframe into final exportable output
final_data = pd.concat([experience.reset_index(drop=True), education], axis=1)
final_data
```
